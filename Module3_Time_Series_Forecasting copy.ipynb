## Objective
Teach forecasting foundations from classical to advanced machine learning methods.

## Core Models & Equations
- **ARIMA/SARIMAX**: \( y_t = c + \phi_1 y_{t-1} + ... + \epsilon_t + \Theta_1 \epsilon_{t-1} \)
- **ETS (Holt-Winters)**: \( y_t = (Level + Trend) * Seasonality + Error \)
- **STL Decomposition**: \( y_t = T_t + S_t + R_t \)
- **Prophet**: \( y(t) = g(t) + s(t) + h(t) + \epsilon_t \)
- **LSTM**: sequence model with memory cells capturing temporal dependencies
- **State-Space (UCM)**: latent structure with transition and observation equations
- **XGBoost**: tree-based regression with lag features
- **AutoML**: automated model selection and stacking

```python
# 
# Module 3: Time Series Forecasting - 
# ==============================================================
# This notebook contains:
#  - Classical methods (ARIMA/SARIMAX)
#  - Exponential Smoothing (Holt-Winters / ETS)
#  - Decomposition-based (STL)
#  - Additive/trend methods (Prophet)
#  - Deep Learning (LSTM/GRU)
#  - Probabilistic (BSTS concept + Monte Carlo sampling)
#  - Multivariate (VAR, DFM sketch)
#  - Tree-based (XGBoost/LightGBM with lag features)
#  - AutoML (PyCaret / AutoTS)
#  - State Space / Unobserved Components (Structural Time Series)
#  - Full Out-Of-Time (OOT) validation for every model with Actual vs Predicted plots
#  - Detailed technical explanations and equations for each sub-topic

# NOTE TO STUDENTS: This file is intended to be run in a Jupyter environment. Each section
# contains explanatory markdown (as comments) followed by runnable code. Replace remote CSV paths
# with local copies if running without internet access.

# ------------------------------------------------------
# 1. Imports & Setup
# ------------------------------------------------------
import warnings
warnings.filterwarnings('ignore')

import os, sys
import pandas as pd
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta

from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.seasonal import STL
from statsmodels.tsa.statespace.structural import UnobservedComponents
from statsmodels.tsa.api import VAR

from prophet import Prophet

#import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.utils.validation import validate_data
#import name 'validate_data' from 'sklearn.utils.validation'
from darts import TimeSeries
from darts.models import RNNModel
from darts.metrics import mae
import xgboost
from xgboost import XGBRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
#import StatsForecast
import optuna
import pycaret
from pycaret.time_series import *

import plotly.graph_objects as go
from properscoring import crps_ensemble
```

```python

# ------------------------------------------------------
# Utility functions: metrics, OOT split, plotting
# ------------------------------------------------------

def train_test_split_oot(series, test_periods=12):
    """Return train series and out-of-time (OOT) test series by a number of periods."""
    train = series[:-test_periods]
    test = series[-test_periods:]
    return train, test


def plot_actual_vs_pred(train, test, forecast, ci_lower=None, ci_upper=None, title='Actual vs Predicted'):
    plt.figure(figsize=(12,5))
    plt.plot(train.index, train, label='Train')
    plt.plot(test.index, test, label='Actual (OOT)', marker='o')
    plt.plot(forecast.index, forecast, label='Forecast', marker='o')
    if ci_lower is not None and ci_upper is not None:
        plt.fill_between(forecast.index, ci_lower, ci_upper, color='lightgray', alpha=0.5, label='Prediction Interval')
    plt.legend(); plt.title(title)
    plt.show()


def compute_metrics(y_true, y_pred):
    mae_val = mean_absolute_error(y_true, y_pred)
    rmse_val = mean_squared_error(y_true, y_pred, squared=False)
    return {'MAE': mae_val, 'RMSE': rmse_val}


def coverage_percentage(y_true, lower, upper):
    inside = ((y_true >= lower) & (y_true <= upper)).sum()
    return inside / len(y_true)
```

```python
# ------------------------------------------------------
# 2. Load Example Data (AirPassengers) & Describe
# ------------------------------------------------------
# The airline passengers dataset is monthly from 1949-1960 and exhibits trend & seasonality.

y = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv",
    parse_dates=['Month'], index_col='Month'
).squeeze()

y.name = 'Passengers'
print('Series frequency (inferred):', pd.infer_freq(y.index))
print('Start:', y.index.min(), 'End:', y.index.max())

y.plot(title='Monthly AirPassengers (1949-1960)', figsize=(12,4)); plt.show()

# Perform Out-of-Time split (OOT) for evaluation: last 24 months held-out
train, test = train_test_split_oot(y, test_periods=24)
print('Train range:', train.index.min(), 'to', train.index.max())
print('Test range:', test.index.min(), 'to', test.index.max())
```

```python
# ------------------------------------------------------
# 3. Classical: ARIMA / SARIMAX (Theory + Code + OOT)
# ------------------------------------------------------
# Theory & Equations (ARIMA):
# ARIMA(p,d,q):
#   y_t' = c + phi_1 y_{t-1}' + ... + phi_p y_{t-p}' + theta_1 e_{t-1} + ... + theta_q e_{t-q} + e_t
# where y_t' indicates differenced series (d differences).
# Seasonal SARIMA adds seasonal polynomials with period s: (P,D,Q)_s

# Fit SARIMAX on training set and forecast OOT horizon
sarima_order = (1,1,1)
seasonal_order = (1,1,1,12)
model_sarimax = SARIMAX(train, order=sarima_order, seasonal_order=seasonal_order,
                        enforce_stationarity=False, enforce_invertibility=False)
res_sarimax = model_sarimax.fit(disp=False)
print(res_sarimax.summary())

# Forecast OOT horizon (24 months)
steps = len(test)
fc_sarimax = res_sarimax.get_forecast(steps=steps)
mean_sarimax = fc_sarimax.predicted_mean
ci_sarimax = fc_sarimax.conf_int()

# Plot Actual vs Predicted with CI
plot_actual_vs_pred(train, test, mean_sarimax, ci_sarimax.iloc[:,0], ci_sarimax.iloc[:,1],
                    title='SARIMAX OOT Forecast vs Actual')

# Metrics & Coverage
metrics_sarimax = compute_metrics(test.values, mean_sarimax.values)
coverage_sarimax = coverage_percentage(test.values, ci_sarimax.iloc[:,0].values, ci_sarimax.iloc[:,1].values)
print('SARIMAX metrics:', metrics_sarimax)
print(f'SARIMAX 95% CI coverage: {coverage_sarimax:.2%}')
```

```python
# ------------------------------------------------------
# 4. ETS (Error-Trend-Seasonal) / Holt-Winters (Theory + Code + OOT)
# ------------------------------------------------------
# Theory: ETS models are a family defined by error (additive/multiplicative),
# trend (none/additive/multiplicative), and seasonality (none/additive/multiplicative).
# One popular form (additive errors) can be written as:
#   y_t = l_{t-1} + b_{t-1} + s_{t-m} + e_t
#   l_t = l_{t-1} + b_{t-1} + alpha * e_t
#   b_t = b_{t-1} + beta * e_t
#   s_t = s_{t-m} + gamma * e_t
# where l=level, b=trend, s=seasonal, m=seasonal_period

# Fit additive Holt-Winters (ETS A,A,A)
ets_model = ExponentialSmoothing(train, trend='add', seasonal='add', seasonal_periods=12)
res_ets = nets_model.fit(optimized=True)
fc_ets = res_ets.forecast(steps)

# Statsmodels' ExponentialSmoothing doesn't return CI by default; we can use naive residual-based
residuals = res_ets.resid
sigma = residuals.std()
ci_lower_ets = fc_ets - 1.96*sigma
ci_upper_ets = fc_ets + 1.96*sigma

plot_actual_vs_pred(train, test, fc_ets, ci_lower_ets, ci_upper_ets, title='ETS (Holt-Winters) OOT Forecast')
print('ETS metrics:', compute_metrics(test.values, fc_ets.values))
print('ETS approximate 95% CI coverage:', coverage_percentage(test.values, ci_lower_ets.values, ci_upper_ets.values))

```

```python
# ------------------------------------------------------
# 5. Decomposition-Based Methods (STL) - Use as Preprocessing + Visualization
# ------------------------------------------------------
# Theory: STL (Seasonal and Trend decomposition using Loess) decomposes series into
# trend + seasonal + remainder. It is robust and useful for diagnostics but not directly a forecasting model.
stl = STL(train, seasonal=13)
res_stl = stl.fit()
res_stl.plot(); plt.show()

# Use STL to remove seasonality and feed deseasonalized series to ARIMA as an example
seasonal_component = res_stl.seasonal
deseasonalized = train - seasonal_component

model_arima_on_deseason = SARIMAX(deseasonalized, order=(1,1,1))
res_arima_deseason = model_arima_on_deseason.fit(disp=False)
fc_deseason = res_arima_deseason.get_forecast(steps=steps).predicted_mean
# add back seasonal pattern by repeating last seasonal cycle values
seasonal_last = seasonal_component[-12:]
seasonal_future = np.tile(seasonal_last.values, int(np.ceil(steps/12)))[:steps]
fc_arima_from_deseason = fc_deseason + seasonal_future
plot_actual_vs_pred(train, test, pd.Series(fc_arima_from_deseason, index=test.index), None, None,
                    title='ARIMA on Deseasonalized (STL) Forecast')
print('ARIMA on deseason metrics:', compute_metrics(test.values, fc_arima_from_deseason.values))

```

```python



# ------------------------------------------------------
# 6. Additive / Trend-based: Prophet (Theory + Code + OOT)
# ------------------------------------------------------
# Theory: Prophet models y(t) = g(t) + s(t) + h(t) + epsilon
# where g(t) is a piecewise-linear or logistic trend with automatic changepoint detection,
# s(t) represents periodic seasonality, and h(t) are holiday effects.

df_train = train.reset_index().rename(columns={'Month':'ds','Passengers':'y'})
df_test = test.reset_index().rename(columns={'Month':'ds','Passengers':'y'})

m = Prophet(yearly_seasonality=True, weekly_seasonality=False)
# we could add holidays; here we'll keep default
m.fit(df_train)
future = m.make_future_dataframe(periods=steps, freq='M')
forecast_prophet = m.predict(future)
# Extract our OOT window
fc_prophet_oot = forecast_prophet.set_index('ds').loc[test.index]

plot_actual_vs_pred(train, test, fc_prophet_oot['yhat'], fc_prophet_oot['yhat_lower'], fc_prophet_oot['yhat_upper'],
                    title='Prophet OOT Forecast vs Actual')
print('Prophet metrics:', compute_metrics(test.values, fc_prophet_oot['yhat'].values))
print('Prophet 95% CI coverage:', coverage_percentage(test.values, fc_prophet_oot['yhat_lower'].values, fc_prophet_oot['yhat_upper'].values))

# ------------------------------------------------------
# 7. State Space Models / Unobserved Components (Theory + Code + OOT)
# ------------------------------------------------------
# Theory: State Space models express time series via Observation eqn and State eqn:
# Observation: y_t = Z_t * alpha_t + epsilon_t   (epsilon_t ~ N(0, H_t))
# State:       alpha_{t+1} = T_t * alpha_t + R_t * eta_t  (eta_t ~ N(0, Q_t))
# UnobservedComponents in statsmodels supports local level, trend, seasonal and cycle components.

# Fit structural model with local level, local linear trend, and seasonal components
ucm = UnobservedComponents(endog=train, level='local linear trend', seasonal=12)
res_ucm = ucm.fit(disp=False)
print(res_ucm.summary())

# Forecast using the state space model
fc_ucm = res_ucm.get_forecast(steps=steps)
mean_ucm = fc_ucm.predicted_mean
ci_ucm = fc_ucm.conf_int()
plot_actual_vs_pred(train, test, mean_ucm, ci_ucm.iloc[:,0], ci_ucm.iloc[:,1], title='State-Space (UCM) OOT Forecast')
print('UCM metrics:', compute_metrics(test.values, mean_ucm.values))
print('UCM 95% CI coverage:', coverage_percentage(test.values, ci_ucm.iloc[:,0].values, ci_ucm.iloc[:,1].values))
```

```python

# ------------------------------------------------------
# 8. Deep Learning: LSTM/GRU (Theory + Code + OOT)
# ------------------------------------------------------
# Theory: RNNs like LSTM (Long Short-Term Memory) and GRU capture sequential dependencies
# using gating mechanisms. They operate on windows of past timesteps as input.
# Loss is typically MSE; for probabilistic outputs you can predict quantiles or ensembles.

# Prepare supervised learning dataset using sliding windows across entire series,
# but ensure training uses only train portion for model fitting.

def make_windows(series, window=12):
    data = series.values.reshape(-1,1)
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(data)
    X, y_ = [], []
    for i in range(len(data_scaled)-window):
        X.append(data_scaled[i:i+window])
        y_.append(data_scaled[i+window])
    return np.array(X), np.array(y_), scaler

window = 12
X_all, y_all, scaler_dl = make_windows(y, window=window)
# determine split index corresponding to train length
split_idx = len(train) - window
X_train_dl = X_all[:split_idx]
y_train_dl = y_all[:split_idx]
X_test_dl = X_all[split_idx:split_idx+len(test)]
y_test_dl = y_all[split_idx:split_idx+len(test)]

# Build simple LSTM
tf.keras.backend.clear_session()
model_dl = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(window,1)),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(1)
])
model_dl.compile(optimizer='adam', loss='mse')
model_dl.fit(X_train_dl, y_train_dl, epochs=50, batch_size=8, verbose=0)

# Forecast OOT horizon using recursive strategy
last_input = X_all[split_idx-1].reshape(1,window,1)
preds_scaled = []
for _ in range(len(test)):
    next_scaled = model_dl.predict(last_input, verbose=0)
    preds_scaled.append(next_scaled[0,0])
    last_input = np.append(last_input[:,1:,:], [[next_scaled]], axis=1)

preds_dl = scaler_dl.inverse_transform(np.array(preds_scaled).reshape(-1,1)).flatten()
fc_index = test.index
plot_actual_vs_pred(train, test, pd.Series(preds_dl, index=fc_index), None, None, title='LSTM OOT Forecast vs Actual')
print('LSTM metrics:', compute_metrics(test.values, preds_dl))

# ------------------------------------------------------
# 9. Probabilistic: BSTS / Gaussian Process (Concept + Monte Carlo)
# ------------------------------------------------------
# Full BSTS implementations are most mature in R (bsts). In Python, we can build
# Bayesian structural models using PyMC. Here we illustrate Monte Carlo predictive sampling
# from residuals to produce an ensemble for CRPS evaluation.

# Create an ensemble of forecasts by bootstrapping residuals from ETS model
n_samples = 200
ensemble_preds = np.zeros((len(test), n_samples))
for i in range(n_samples):
    # simulate residuals by sampling with replacement and adding to point forecast
    sampled_resids = np.random.choice(res_ets.resid, size=len(test), replace=True)
    ensemble_preds[:, i] = fc_ets.values + sampled_resids

# Compute CRPS for the ensemble vs actual
crps_vals = crps_ensemble(test.values, ensemble_preds)
print('Mean CRPS for ETS-residual-bootstrap ensemble:', crps_vals.mean())

# Visualize probabilistic bands (percentiles)
lower_pct = np.percentile(ensemble_preds, 2.5, axis=1)
upper_pct = np.percentile(ensemble_preds, 97.5, axis=1)
median_pred = np.percentile(ensemble_preds, 50, axis=1)
plot_actual_vs_pred(train, test, pd.Series(median_pred, index=test.index), lower_pct, upper_pct, title='Probabilistic Forecast (Residual Bootstraps)')
print('Coverage of 95% empirical bands:', coverage_percentage(test.values, lower_pct, upper_pct))

# ------------------------------------------------------
# 10. Multivariate: VAR (Theory + Code + OOT)
# ------------------------------------------------------
# Build a toy multivariate dataset: passenger series and a lagged copy (as an example)
df_var = pd.DataFrame({'y1': y, 'y2': y.shift(1).fillna(method='bfill')})
train_var = df_var.iloc[:len(train)]
test_var = df_var.iloc[len(train):]

model_var = VAR(train_var)
res_var = model_var.fit(maxlags=12, ic='aic')
print('Selected lag order (VAR):', res_var.k_ar)

# Forecast VAR for steps ahead
fc_var = res_var.forecast(train_var.values[-res_var.k_ar:], steps=len(test_var))
fc_var_df = pd.DataFrame(fc_var, index=test_var.index, columns=train_var.columns)

# Plot first variable actual vs predicted
plot_actual_vs_pred(train, test, fc_var_df['y1'], None, None, title='VAR OOT Forecast (y1 vs Actual)')
print('VAR metrics (y1):', compute_metrics(test.values, fc_var_df['y1'].values))

# ------------------------------------------------------
# 11. Tree-Based Models (XGBoost with lags) - Theory + Code + OOT
# ------------------------------------------------------
# Create lag features and time features (month, trend)
lag_df = pd.DataFrame({'y': y})
for lag in range(1,25):
    lag_df[f'lag_{lag}'] = lag_df['y'].shift(lag)
lag_df['month'] = lag_df.index.month
lag_df['time_idx'] = np.arange(len(lag_df))
lag_df.dropna(inplace=True)

# Split into train/test by index preserving time-order
train_lag = lag_df.iloc[:len(train)]
test_lag = lag_df.iloc[len(train):]
X_train = train_lag.drop('y', axis=1)
y_train = train_lag['y']
X_test = test_lag.drop('y', axis=1)
y_test = test_lag['y']

xgb = XGBRegressor(n_estimators=500, objective='reg:squarederror')
xgb.fit(X_train, y_train)
fc_xgb = xgb.predict(X_test)
plot_actual_vs_pred(train, test, pd.Series(fc_xgb, index=test.index), None, None, title='XGBoost (lags) OOT Forecast')
print('XGBoost metrics:', compute_metrics(test.values, fc_xgb))

# ------------------------------------------------------
# 12. AutoML: PyCaret (Sketch) - Use for Baseline Comparison
# ------------------------------------------------------
# PyCaret's time series module automates model comparison and ensembling. This is left
# as a commented optional block since it requires interactive setup and may be slow.

# from pycaret.time_series import *
# s = setup(data=y, fh=len(test))
# best = compare_models()  # WARNING: may download models and is time-consuming
# print(best)

# ------------------------------------------------------
# 13. Ensembling (Stacking) - Combine diverse models
# ------------------------------------------------------
# Construct a meta-dataset using OOT predictions from SARIMAX, ETS, Prophet, LSTM, XGBoost
pred_sarimax = mean_sarimax.reindex(test.index)
pred_ets = fc_ets.reindex(test.index)
pred_prophet = fc_prophet_oot['yhat']
pred_lstm = pd.Series(preds_dl, index=test.index)
pred_xgb = pd.Series(fc_xgb, index=test.index)

meta_X = pd.DataFrame({
    'sarimax': pred_sarimax,
    'ets': pred_ets,
    'prophet': pred_prophet,
    'lstm': pred_lstm,
    'xgb': pred_xgb
}).fillna(method='ffill')

# Train simple Ridge meta-learner on the OOT predictions (here we use same OOT as meta-training for demo)
meta = Ridge(alpha=1.0)
meta.fit(meta_X, test.values)
ensemble_preds = meta.predict(meta_X)

plot_actual_vs_pred(train, test, pd.Series(ensemble_preds, index=test.index), None, None, title='Ensemble (Stacking) OOT Forecast')
print('Ensemble metrics:', compute_metrics(test.values, ensemble_preds))

```
