## **Conjoint Analysis: Detailed Explanation with Example**

Conjoint analysis is a statistical technique used in market research to understand customer preferences and how they make trade-offs between different product attributes or features. It helps identify the optimal combination of product features that will appeal to consumers, based on their preferences and priorities.

### Key Concepts:

1. **Attributes and Levels**: In conjoint analysis, products or services are broken down into **attributes** (features or characteristics), and each attribute has different **levels** (specific options or variations of the attribute). For example, in a smartphone, attributes could include screen size, battery life, camera quality, and price, and their levels could vary, such as "5.5-inch," "4000 mAh," "12 MP camera," and "under $500."

2. **Trade-offs**: Consumers usually cannot have everything they want in a product; thus, conjoint analysis focuses on understanding the **trade-offs** consumers are willing to make between different attributes. This helps in understanding which attributes are most important in decision-making.

3. **Utility**: Conjoint analysis derives **part-worth utilities**, which represent the perceived value of each level of an attribute. These utilities indicate how much consumers value a particular feature. The higher the utility, the more important or attractive the feature is to the consumer.

4. **Choice Modeling**: The technique simulates real-world purchasing decisions by asking participants to choose between different product profiles (combinations of attributes and their levels). Based on these choices, conjoint analysis estimates how consumers value the various attributes and their levels.

### Steps in Conjoint Analysis:

1. **Define the Attributes and Levels**: The first step is identifying the key attributes and their respective levels. For example, when designing a car, you might consider attributes like color, engine type, and price, with levels like "red" or "blue" for color, "electric" or "gas" for engine type, and price ranges.

2. **Create Product Profiles**: Next, you generate different combinations of attribute levels to create **product profiles** (hypothetical products). Each profile represents a different combination of attributes.

3. **Conduct a Survey or Experiment**: Participants (usually potential customers) are shown various product profiles and asked to choose their preferred option. This can be done through surveys, interviews, or online platforms. You can ask them to rank, rate, or choose between different options.

4. **Statistical Analysis**: Once the data is collected, **statistical models** (such as regression analysis or maximum likelihood estimation) are applied to determine the utility values (part-worth utilities) for each attribute level. This analysis helps to reveal how much each attribute contributes to the decision-making process.

5. **Interpret Results and Predict Preferences**: The results are then interpreted to understand consumer preferences, and simulations can predict which product configurations will be most successful in the market.

### Example of Conjoint Analysis:

Let’s say you are a company designing a new **smartphone**, and you want to understand what features customers care about most in their buying decisions.

#### Step 1: Define Attributes and Levels

- **Screen Size**: 5.5 inches, 6.0 inches, 6.5 inches
- **Battery Life**: 3000 mAh, 4000 mAh, 5000 mAh
- **Camera Quality**: 8 MP, 12 MP, 16 MP
- **Price**: $300, $500, $700

#### Step 2: Create Product Profiles

Using these attributes and levels, you can create product profiles, such as:

- **Profile 1**: Screen size = 6.0 inches, Battery life = 4000 mAh, Camera quality = 12 MP, Price = $500
  
- **Profile 2**: Screen size = 5.5 inches, Battery life = 3000 mAh, Camera quality = 8 MP, Price = $300

- **Profile 3**: Screen size = 6.5 inches, Battery life = 5000 mAh, Camera quality = 16 MP, Price = $700

#### Step 3: Collect Data

You then show these profiles to potential customers and ask them to rank or choose which smartphone they prefer.

#### Step 4: Analyze Data

After collecting customer preferences, you would use statistical models to analyze the data and determine the **part-worth utilities** for each attribute level. For example:

- **Screen Size Utility**: 6.5 inches might have the highest utility, followed by 6.0 inches and then 5.5 inches.
- **Battery Life Utility**: Customers may value 5000 mAh the most, followed by 4000 mAh and then 3000 mAh.
- **Camera Quality Utility**: 16 MP might be the highest, followed by 12 MP, and then 8 MP.
- **Price Utility**: The lower the price, the higher the utility, with $300 having the highest utility.

#### Step 5: Interpretation and Decision Making

From the results, you can interpret that customers prioritize **battery life**, **camera quality**, and **screen size** over **price**. This insight helps you decide which combination of attributes to focus on in your product. For example, you may choose to offer a smartphone with a **6.5-inch screen**, **5000 mAh battery**, **16 MP camera**, and a **price of $500** (because this offers a good balance of preferred attributes without being too expensive).

You can also use **market simulations** to predict how different product offerings would perform in the real world. For instance, if you offer a model at $700 with a high battery life and camera quality, you might see a drop in preference due to its high price.

### Benefits of Conjoint Analysis:

2. **Quantifies Consumer Preferences**: It allows companies to understand how consumers value specific attributes and make trade-offs.
3. **Pricing Strategy**: It helps businesses determine the optimal price points based on consumer willingness to pay for different features.
4. **Market Segmentation**: It can uncover different customer segments with distinct preferences, helping tailor products to various groups.

### **Practical example** of performing **Conjoint Analysis** using **Python**. 

We'll simulate a simple conjoint analysis scenario for a **smartphone market** with the following attributes:

- **Screen Size** (5.5 inches, 6.0 inches, 6.5 inches)
- **Battery Life** (3000 mAh, 4000 mAh, 5000 mAh)
- **Camera Quality** (8 MP, 12 MP, 16 MP)
- **Price** ($300, $500, $700)

We will then create product profiles, simulate consumer preferences, and use **regression analysis** to derive part-worth utilities for each attribute level.

### Steps:
1. **Generate Product Profiles**: We will combine each attribute level to generate possible product profiles.
2. **Simulate Consumer Preferences**: We'll simulate consumer preferences by creating a hypothetical survey, asking consumers to choose between different product profiles based on their preferences.
3. **Estimate Part-Worth Utilities**: Use **linear regression** to estimate part-worth utilities from the data.

Let's implement this step by step.

### Python Code Implementation:

````python
import itertools
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

np.random.seed(0)

# Attributes and levels
attributes = {
    "Screen Size": [5.5, 6.0, 6.5],
    "Battery Life": [3000, 4000, 5000],
    "Camera Quality": [8, 12, 16],
    "Price": [300, 500, 700]
}

# Full factorial profiles
product_profiles = list(itertools.product(
    attributes["Screen Size"],
    attributes["Battery Life"],
    attributes["Camera Quality"],
    attributes["Price"]
))
profiles_df = pd.DataFrame(product_profiles,
                           columns=["Screen Size", "Battery Life", "Camera Quality", "Price"])
profiles_df["Price_k"] = profiles_df["Price"] / 1000.0  # price in $1000

# Synthetic utility (ground truth)
def generate_preferences(row):
    screen_size_utility = {5.5: 2, 6.0: 3, 6.5: 4}
    battery_utility = {3000: 1, 4000: 3, 5000: 5}
    camera_utility = {8: 1, 12: 3, 16: 5}
    price_utility = {300: 5, 500: 3, 700: 1}
    return (screen_size_utility[row["Screen Size"]]
            + battery_utility[row["Battery Life"]]
            + camera_utility[row["Camera Quality"]]
            + price_utility[row["Price"]])

profiles_df["Utility"] = profiles_df.apply(generate_preferences, axis=1)

# --- Important: ensure explicit categorical ordering (guarantee reference level) ---
for attr in ["Screen Size", "Battery Life", "Camera Quality"]:
    # create categorical with levels in the order we want (first becomes reference)
    levels = [str(x) for x in attributes[attr]]
    profiles_df[attr] = profiles_df[attr].astype(str)  # ensure string
    profiles_df[attr] = pd.Categorical(profiles_df[attr], categories=levels, ordered=True)

# One-hot encode categorical attributes (drop_first=True -> reference = first category)
X_cats = pd.get_dummies(profiles_df[["Screen Size", "Battery Life", "Camera Quality"]], drop_first=True)
X = pd.concat([X_cats, profiles_df[["Price_k"]]], axis=1)
y = profiles_df["Utility"]

# Fit linear regression
model = LinearRegression().fit(X, y)
intercept = model.intercept_
price_coef_per_1000 = model.coef_[X.columns.get_loc("Price_k")]

# Recover part-worths: reference levels set to 0
part_worths = []
for attr in ["Screen Size", "Battery Life", "Camera Quality"]:
    levels = [str(x) for x in attributes[attr]]
    ref = float(levels[0])  # reference (first)
    part_worths.append({"Attribute": attr, "Level": ref, "PartWorth": 0.0})
    for lvl in levels[1:]:
        # expected dummy column name format: "Attribute_level"
        colname = f"{attr}_{lvl}"
        pw = model.coef_[X.columns.get_loc(colname)]
        part_worths.append({"Attribute": attr, "Level": float(lvl), "PartWorth": pw})

part_worths_df = pd.DataFrame(part_worths)

# Compute WTP (in dollars) relative to each reference
wtp_rows = []
for attr in ["Screen Size", "Battery Life", "Camera Quality"]:
    grp = part_worths_df[part_worths_df["Attribute"] == attr].set_index("Level")["PartWorth"].to_dict()
    ref = list(grp.keys())[0]
    for lvl, pw in grp.items():
        delta = pw - grp[ref]
        wtp = (delta / (-price_coef_per_1000)) * 1000.0
        wtp_rows.append({"Attribute": attr, "Level": lvl, "PartWorth": pw, "WTP_dollars": wtp})
wtp_df = pd.DataFrame(wtp_rows)

# Predict for new profiles (safe alignment)
new_profiles = pd.DataFrame({
    "Screen Size": [5.5, 6.0, 6.5],
    "Battery Life": [3000, 4000, 5000],
    "Camera Quality": [8, 12, 16],
    "Price": [300, 500, 700]
})
new_profiles["Price_k"] = new_profiles["Price"] / 1000.0
for attr in ["Screen Size", "Battery Life", "Camera Quality"]:
    new_profiles[attr] = new_profiles[attr].astype(str)
X_new_cats = pd.get_dummies(new_profiles[["Screen Size", "Battery Life", "Camera Quality"]], drop_first=True)
X_new_cats = X_new_cats.reindex(columns=X_cats.columns, fill_value=0)  # align columns
X_new = pd.concat([X_new_cats, new_profiles[["Price_k"]]], axis=1)
new_profiles["Predicted_Utility"] = model.predict(X_new)

# Print results
print("Intercept:", intercept)
print("\nPart-worths:")
print(part_worths_df)
print("\nPrice coefficient (per $1000):", price_coef_per_1000)
print("\nWTP table (dollars):")
print(wtp_df)
print("\nPredicted utilities for new profiles:")
print(new_profiles)

# Plot part-worths (one plot)
fig, ax = plt.subplots(figsize=(8,4))
for attr, grp in part_worths_df.groupby("Attribute"):
    ax.plot(grp["Level"].astype(str), grp["PartWorth"], marker='o', label=attr)
ax.set_title("Estimated Part-Worths by Attribute (reference level = 0)")
ax.set_ylabel("Part-Worth Utility")
ax.set_xlabel("Attribute Level")
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# Save as hb_conjoint.py and run with: python hb_conjoint.py
import itertools, numpy as np, pandas as pd
import pymc as pm
import arviz as az

np.random.seed(2)

# 1) Define attributes & profiles
attributes = {
    "Screen Size": [5.5, 6.0, 6.5],
    "Battery Life": [3000, 4000, 5000],
    "Camera Quality": [8, 12, 16],
    "Price": [18000, 22000, 26000]
}
profiles = list(itertools.product(attributes["Screen Size"], attributes["Battery Life"], attributes["Camera Quality"], attributes["Price"]))
profiles_df = pd.DataFrame(profiles, columns=["Screen Size","Battery Life","Camera Quality","Price"]).reset_index().rename(columns={"index":"profile_id"})
profiles_df["Price_k"] = profiles_df["Price"]/1000.0

# 2) Create design: respondents × paired tasks
n_respondents = 120
tasks_per_respondent = 6

def sample_pairs(n_pairs, n_profiles):
    idx = np.random.choice(n_profiles, size=(n_pairs,2), replace=True)
    for i in range(n_pairs):
        if idx[i,0]==idx[i,1]:
            idx[i,1] = (idx[i,1]+1) % n_profiles
    return idx

design_records = []
for r in range(n_respondents):
    pairs = sample_pairs(tasks_per_respondent, len(profiles_df))
    for t in range(tasks_per_respondent):
        design_records.append({"resp_id": r, "task_id": t, "L_profile": int(pairs[t,0]), "R_profile": int(pairs[t,1])})
design = pd.DataFrame(design_records)

# merge profile attributes to design (safe merges on profile_id)
#design = design.merge(profiles_df.add_prefix("L_"), left_on="L_profile", right_on="L_profile", how="left")
#design = design.merge(profiles_df.add_prefix("R_"), left_on="R_profile", right_on="R_profile", how="left")

# design has columns: ['resp_id','task_id','L_profile','R_profile']

# Merge LEFT profile attributes (profiles_df.profile_id -> L_profile_id after add_prefix)
design = design.merge(
    profiles_df.add_prefix("L_"),
    left_on="L_profile",
    right_on="L_profile_id",
    how="left"
)

# Merge RIGHT profile attributes
design = design.merge(
    profiles_df.add_prefix("R_"),
    left_on="R_profile",
    right_on="R_profile_id",
    how="left"
)

# Optional: drop the duplicate id columns introduced by the merges
for col in ["L_profile_id", "R_profile_id"]:
    if col in design.columns:
        design = design.drop(columns=[col])

print("design.columns:", design.columns.tolist())
print("profiles_df.columns:", profiles_df.columns.tolist())
````

````python
# 3) Effects-coding helper (2 columns per 3-level attribute)
def effects_vec(screen, battery, camera):
    vec = []
    # screen: columns for 5.5, 6.0 ; 6.5 is implicit with -1 in both
    for lvl in [5.5, 6.0]:
        vec.append(1 if float(screen)==lvl else (-1 if float(screen)==6.5 else 0))
    # battery: 3000, 4000 ; 5000 implicit
    for lvl in [3000, 4000]:
        vec.append(1 if float(battery)==lvl else (-1 if float(battery)==5000 else 0))
    # camera: 8, 12 ; 16 implicit
    for lvl in [8, 12]:
        vec.append(1 if float(camera)==lvl else (-1 if float(camera)==16 else 0))
    return np.array(vec)

# 4) Simulate respondent-level "true" betas (for data generation)
mu_vec = np.array([-1.0,0.0,-1.0,0.0,-1.0,0.0])  # population effects-coded means
sigma_true = 0.8
resp_betas = {r: np.random.normal(mu_vec, sigma_true) for r in range(n_respondents)}
resp_price = {r: -10.0 + np.random.normal(0,1.0) for r in range(n_respondents)}

# 5) Simulate choices using random utility + Gumbel noise
choices = []
for i,row in design.iterrows():
    r = int(row["resp_id"])
    Lvec = effects_vec(row["L_Screen Size"], row["L_Battery Life"], row["L_Camera Quality"])
    Rvec = effects_vec(row["R_Screen Size"], row["R_Battery Life"], row["R_Camera Quality"])
    Lu = resp_betas[r].dot(Lvec) + resp_price[r]*(row["L_Price"]/1000.0)
    Ru = resp_betas[r].dot(Rvec) + resp_price[r]*(row["R_Price"]/1000.0)
    Lu_n = Lu + (-np.log(-np.log(np.random.rand())))
    Ru_n = Ru + (-np.log(-np.log(np.random.rand())))
    choices.append(int(Lu_n > Ru_n))
design["L_chosen"] = choices

# 6) Build X_diff and arrays for model
n_obs = len(design); p = 6
X_diff = np.zeros((n_obs, p+1))
resp_idx = design["resp_id"].astype(int).values
y_obs = design["L_chosen"].values
for i,row in design.iterrows():
    Lvec = effects_vec(row["L_Screen Size"], row["L_Battery Life"], row["L_Camera Quality"])
    Rvec = effects_vec(row["R_Screen Size"], row["R_Battery Life"], row["R_Camera Quality"])
    X_diff[i,:p] = (Lvec - Rvec)
    X_diff[i,p] = (row["L_Price"] - row["R_Price"]) / 1000.0

````

````python
# 7) PyMC hierarchical logistic regression
with pm.Model() as hierarchical_model:
    mu = pm.Normal("mu", mu=0.0, sigma=5.0, shape=p+1)
    sigma = pm.HalfNormal("sigma", sigma=2.0, shape=p+1)
    beta = pm.Normal("beta", mu=mu, sigma=sigma, shape=(n_respondents, p+1))
    logits = (beta[resp_idx] * X_diff).sum(axis=1)
    y = pm.Bernoulli("y", logit_p=logits, observed=y_obs)
   # trace = pm.sample(draws=200, tune=200, chains=2, cores=1, random_seed=42)
    if __name__ == "cores=1":
        with hierarchical_model:
            trace = pm.sample(draws=200, tune=200, chains=2, cores=2, random_seed=42)
````

````python
# 8) Inspect posterior
mu_post = trace.posterior["mu"].mean(dim=("chain","draw")).values
price_mu = mu_post[-1]
print("Population posterior mean (mu):", mu_post)
print("Population price coef per $1000 (posterior mean):", price_mu)

# Map mu_post into part-worths (effects-coded -> full levels)
pop_pw = {}
idx = 0
attrs = [("Screen Size",[5.5,6.0,6.5]), ("Battery Life",[3000,4000,5000]), ("Camera Quality",[8,12,16])]
for attr, levels in attrs:
    coefs = mu_post[idx:idx+len(levels)-1]
    last = -coefs.sum()
    mapping = {}
    for lvl, c in zip(levels[:-1], coefs):
        mapping[float(lvl)] = float(c)
    mapping[float(levels[-1])] = float(last)
    pop_pw[attr] = mapping
    idx += len(levels)-1

print("Posterior population part-worths (means):")
for a,m in pop_pw.items():
    print(a, m)
````

### **Explanation of Output**

* `mu` posterior mean ≈ population average part-worths (effects-coded coefficients, then mapped to levels).
* `sigma` shows heterogeneity (higher sigma => more respondent heterogeneity for that coefficient).

Meaning:

* Magnitude tells how strongly price influences preference.

This is central for **WTP (Willingness to Pay)**:

[
WTP = \frac{\Delta \text{partworth}}{-\beta_{price}} \times 1000

#### 4. How to interpret magnitudes (utility scale)

Utilities are on an **interval scale**, meaning:

* Only **rank order** and **differences** matter
* Higher = more preferred
* Differences = strength of preference

Example:

Battery Life difference:
[
5000 - 3000 = 0.969 - (-1.450) ≈ 2.42
]

Camera Quality difference:
[
16 - 8 = 0.9766 - (-1.0389) ≈ 2.01
]

This means:
**Going from 3000→5000 mAh affects choice more than going 8MP→16MP.**

#### 5. **How to interpret in business terms (WTP)**

Use:

[
WTP = \frac{\Delta \text{partworth}}{-\beta_{price}} \times 1000
]

###### Example: Battery Life — 5000 mAh vs 3000 mAh

Δ utility = 2.42
Price coefficient = −3.02

[
WTP = \frac{2.42}{3.02} × 1000 ≈ $802
]

Meaning:

# Short recap & recommendations

* Effects coding gives interpretable part-worths whose **sum = 0** per attribute; price as numeric yields direct **WTP** conversion.
* Bootstrapping yields empirical CIs; with real (noisy) data this becomes important to judge statistical significance of level differences.
* Hierarchical Bayes is the gold standard to capture respondent heterogeneity — use PyMC or dedicated HB conjoint packages for production. The script above is ready to run; increase sampling and check convergence diagnostics (`r_hat` / ESS) before trusting detailed numeric output.
